```python
import numpy as np
import pandas as pd
from datetime import datetime


df = pd.read_csv('data_source/data.log.gz_csv', delimiter=';',
                   names=['ip', 'datetime', 'method', 'gateway', 'protocol', 'status', 'size'])
```

```python
# old_df = df
# df = old_df.copy()
```

```python
df['dt'] = df['datetime'].apply(lambda v: datetime.strptime(v, '%d/%b/%Y:%H:%M:%S %z'))
```

```python
# df['date'] = df['dt'].apply(lambda v: v.date())
```

```python
df['times'] = df['dt'].apply(lambda v: v.replace(second=0, minute=int(v.minute) - int(v.minute) % 10))
```

```python
gf = df.groupby(by=['times']).count()['ip'].reset_index()

gf = gf[gf['times'].apply(lambda v: v.year == 2015 and v.month == 2 and v.day == 6)]
```

```python
gf = gf.set_index('times')
```

```python
gf.plot(kind='line', figsize=(20, 7), grid=True)
```

```python
pivot = pd.pivot_table(df[df['times'].apply(lambda v: v.year == 2015 and v.month == 2 and v.day == 6)], index=['times'], values=['ip'], columns=['status'], aggfunc='count', fill_value=0)
```

```python
pivot.plot(figsize=(20, 8))
```
